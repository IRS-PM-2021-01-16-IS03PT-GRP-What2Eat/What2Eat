```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Shaun\Downloads\NUS Master of Intelligent Systems\Practice Module\Dataset\raw-data_recipe.csv")
```

```python
df500 = df.sample(n=500)
```

```python
#reset the index for easy reference
df500 = df500.reset_index(drop=True)
df500
```

```python
#look into cleaning cooking_directions

df500['cooking_directions'][0]
```

```python
for i in range(len(df500['cooking_directions'])):
    df500['cooking_directions'][i] = df500['cooking_directions'][i].replace("\\n"," ")
    df500['cooking_directions'][i] = df500['cooking_directions'][i].replace("u'"," ")
```

```python
df500
```

```python
#look into cleaning nutritions column

df500['nutritions'][0]
```

```python
for i in range(len(df500['nutritions'])):
    df500['nutritions'][i] = df500['cooking_directions'][i].replace("u'"," ")
```
